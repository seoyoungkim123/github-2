중학교 남학생들의 몸무게에 대한 대푯값 출력

```python
import numpy as np
import pandas as pd
from scipy import stats

x=pd.read_excel("/content/중학생_남자_몸무게.xlsx")

#평균계산
print('mean=',np.average(x),'wn')


#중앙값 계산
print('median=',np.median(x),'wn')


#최빈값 계산
print('mode',stats.mode(x))
```

```python
import numpy as np
import pandas as pd

data=pd.DataFrame(pd.read_excel("/content/외식비.xlsx"))

print('범위range')
print(np.max(data)-np.min(data))
print('wn')

print('분산 variance')
print(np.var(data))
print('wn')

print('표준편차standard deviation')
print(np.std(data))
print('wn')

print('사분위수 편차 quartile deviation')
q25=data.quantile(0.25)
q75=data.quantile(0.75)
IRQ=q75-q25
print(IRQ/2)
```
